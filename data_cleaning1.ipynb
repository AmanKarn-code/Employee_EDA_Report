```python
# Replaced the null value
data['Age']=data['Age'].fillna(data['Age'].median())
data['Salary']=data['Salary'].fillna(data.groupby('Department')['Salary'].transform('median'))
```

```python
data['Phone']=abs(data['Phone']).astype('int64')
```

```python
data['Department_Region']=data['Department_Region'].str.strip()
data['Full_Name']=data['First_Name']+' '+data['Last_Name']
```

```python
# Corrected the datatype
data['Age']=data['Age'].astype('Int64')
data['Phone']=data['Phone'].astype('Float64')
data['Join_Date'] = pd.to_datetime(data['Join_Date'], errors='coerce')
```

```python
data[['Department','Region']]= data["Department_Region"].str.split('-',n=1,expand=True)
data[['Department','Region','Salary']].head
```

```python
data.to_csv("Employee_Dataset.csv" , index=False)
```
